(b) D√©crivez ce que fait `Bidon` en g√©n√©ral.

**R√©ponse :**

Il y a trois √©tapes principales dans l'√©x√©cution de la fonction bidon():

1.   La fonction compare deux √©l√©ments adjacents de la portion du tableau sp√©cifi√©e par les indices $B_I$ et $B_I+1$, et attribue la plus grande valeur √† $w$ et la plus petite √† $x$. (lignes 2 √† 8)

2.   Si les deux √©l√©ments sont les seuls de la portion $(B_S \leq B_I + 1)$, elle retourne ces deux valeurs. (lignes 9 √† 11 $\rightarrow$ *Cas de base*)

3.   Sinon, elle appelle r√©cursivement la fonction `Bidon` sur le reste de la portion du tableau, et compare $w$ et $x$ aux r√©sultats de l'appel r√©cursif pour trouver le maximum et le minimum de la portion. (lignes 12 √† 27)

En regardant les traces ci-dessous, on peut voir que la fonction retourne bien le max ($n_1$) et le min ($n_2$) de Tab.

````python
def Bidon(Tab, BI, BS):

    # Comparaison des deux premiers √©l√©ments de la portion du tableau
    if Tab[BI] >= Tab[BI + 1]:
        w = Tab[BI]
        x = Tab[BI + 1]

    else:
        w = Tab[BI + 1]
        x = Tab[BI]

    print("w :", w)
    print("x :", x)

    # Cas de base : Comparer BS et BI + 1 -> si BS est plus petit ou √©gal √† BI + 1, assigner n1 et n2
    if BS <= BI + 1:
        print("Cas de base retrouv√©...")
        n1 = w
        n2 = x

    else:
        # Sinon, on fait un appel r√©cursif pour traiter le reste du tableau
        print("Appel r√©cursif...")
        y, z = Bidon(Tab, BI + 2, BS) # On avance BI et on compare les deux prochains chiffres

        print("y :", y)
        print("z :", z)

        # Calcul le maximum et minimum du tableau respectivement
        print("Calcul du maximum (n1) et minimum (n2) du tableau...")
        if w >= y:
            n1 = w
        else:
            n1 = y

        if x <= z:
            n2 = x
        else:
            n2 = z

        print("n1: ", n1)
        print("n2: ", n2)

    # Retourner les deux entiers n1 et n2
    return n1, n2
````

## **Exercice 3**

Soit $Tab$ un tableau de nombres entiers tous diff√©rents les uns des autres dont les indices sont compris entre 1 et $n$. On suppose que les √©l√©ments de $Tab$ sont en ordre croissant et on d√©sire r√©soudre le probl√®me suivant :

<div align="center">

 *Existe-t-il un indice $i$ tel que $Tab[i] = i$?*

</div>

√âcrivez une fonction bool√©enne qui retourne la valeur VRAI si et seulement si un indice $i$ v√©rifiant le probl√®me ci-dessus existe dans le tableau $Tab$. Votre fonction doit prendre un temps proportionnel √† $\log(n)$ dans le pire des cas.

**R√©ponse :**
&nbsp;

L'algorithme qui v√©rifie s'il existe un point fixe dans un tableau tri√© (c'est-√†-dire une position $i$ telle que $T[i]=i$) en utilisant une approche r√©cursive de recherche dichotomique.

Il y a quatre √©tapes principales dans l'√©x√©cution de la fonction :

1.	**Initialisation des bornes** : L'algorithme commence avec des bornes $B_I$ (borne inf√©rieure) et $B_S$ (borne sup√©rieure), qui repr√©sentent les indices extr√™mes du tableau √† examiner.
2.	**Condition d'arr√™t** (lignes 2-3): Si $B_S < B_I$, cela signifie que la recherche est termin√©e sans avoir trouv√© de point fixe, et donc l'algorithme retourne FAUX.
3.	**Calcul du milieu du tableau** (ligne 5) : L'algorithme calcule l'indice du milieu $m$.
4.	**Comparaisons et appels r√©cursif :**
  *   Si $Tab[m]= m$, cela signifie qu'un point fixe est trouv√©, et l'algorithme retourne VRAI. (lignes 6-7)
  *   Si $Tab[m] < m$, cela signifie que le point fixe potentiel doit √™tre √† droite de $m$, donc l'algorithme continue la recherche dans la partie droite. (lignes 8-9)
  * Si $Tab[m] > m$, cela signifie que le point fixe potentiel doit √™tre √† gauche de $m$, donc l'algorithme continue la recherche dans la partie gauche. (lignes 10-11)

*Pseudocode*

```
1: fonction pointFixe(Tab, BI, BS)
      Entr√©e:
        Tab : tableau index√© de 1 √† n
        BI : nombre entier - la borne inf√©rieure (indice de d√©but)
        BS : nombre entier - la borne sup√©rieure (indice de fin)
      Retour:
        Vrai / Faux : Bool√©en

2:     si BS < BI alors
3:         retourner Faux  ‚ñ∑  Si la borne sup√©rieure est inf√©rieure √† la borne inf√©rieure, il n'y a pas de point fixe
4:     sinon
5:         m ‚Üê (BI + BS) √∑ 2  ‚ñ∑  Calcul de l'indice du milieu
6:         si Tab[m] = m alors
7:             retourner Vrai  ‚ñ∑  Si l'√©l√©ment au milieu est √©gal √† son indice, alors on a trouv√© un point fixe
8:         sinon si Tab[m] < m alors
9:             retourner pointFixe(Tab, m + 1, BS)  ‚ñ∑  Si l'√©l√©ment est plus petit que l'indice, chercher dans la partie droite
10:        sinon
11:             retourner pointFixe(Tab, BI, m - 1)  ‚ñ∑  Sinon, chercher dans la partie gauche
12:        fin si
13:     fin si

14: fin fonction
```



````python
def pointFixe(Tab, BI, BS):
    """
    D√©termine si un tableau contient un point fixe.

    Param√®tres:
    - Tab : liste d'entiers index√©e de 0 √† n-1
    - BI : entier - borne inf√©rieure (indice de d√©but)
    - BS : entier - borne sup√©rieure (indice de fin)

    Retour:
    - bool : Vrai si un point fixe existe, Faux sinon
    """
    if BS < BI:
        return False  # Si la borne sup√©rieure est inf√©rieure √† la borne inf√©rieure, pas de point fixe
    else:
        m = (BI + BS) // 2  # Calcul de l'indice du milieu

        if Tab[m] == m:
            return True  # Si l'√©l√©ment au milieu est √©gal √† son indice, on a trouv√© un point fixe
        elif Tab[m] < m:
            return pointFixe(Tab, m + 1, BS)  # Chercher dans la partie droite
        else:
            return pointFixe(Tab, BI, m - 1)  # Chercher dans la partie gauche
````

````python
# Exemple d'utilisation :
tableau = [-15, -10, -5,  3, 7, 10, 15, 20, 25]  # Le point fixe est √† l'indice 3 (Tab[3] = 3)
resultat = pointFixe(tableau, 0, len(tableau) - 1)
print("Point fixe trouv√© :", resultat)
````

## **Exercice 4**

√âcrivez un algorithme efficace pour calculer la racine carr√©e enti√®re d‚Äôun entier $n$ sup√©rieur ou √©gal √† 2. La racine carr√©e enti√®re de $n$ est le plus grand entier positif $r$ tel que $r^2 \leq n$. Par *algorithme efficace*, on entend un algorithme dont le temps d‚Äôex√©cution est proportionnel √† $\log(n)$.

**R√©ponse :**

Encore une fois, on peut utiliser une approche bas√©e sur la recherche dichotomique:  


1.  **Condition d'arr√™t** (lignes 2-3) :
  * Si l'intervalle de recherche ne contient plus que deux √©l√©ments (i.e., $B_S=B_I+1$), alors la racine carr√©e enti√®re est $B_I$. Cela signifie que nous avons trouv√© la plus grande valeur $r$ telle que $r^2 \leq n$.

2. **Division de l'intervalle** (lignes 5-9) :
  *   √Ä chaque √©tape, on divise l'intervalle en deux en calculant le milieu $m$.
  *   Si $m^2 \leq n$, on sait que la racine carr√©e enti√®re se situe √† droite ou √©gale √† $m$.
  *   Si $m ^2 > n$, on sait que la racine carr√©e enti√®re se situe √† gauche de $m$.
3. **R√©cursivit√©**
  * L'algorithme appelle la fonction racine de mani√®re r√©cursive jusqu'√† ce que la condition $B_S=B_I+1$ soit atteinte.


*Pseudocode*

```
1: fonction racine(n, BI, BS : entier naturel) : entier naturel
    Entr√©e:
        n : l'entier dont on veut calculer la racine carr√©e enti√®re
        BI : la borne inf√©rieure de l'intervalle de recherche
        BS : la borne sup√©rieure de l'intervalle de recherche
    Retour:
        entier naturel (racine carr√©e de n)

2:     si BS = BI + 1 alors
3:         retourner BI ‚ñ∑ Si l'intervalle est r√©duit √† sa plus simple expression, BI est la racine carr√©e enti√®re
4:     sinon
5:         milieu ‚Üê (BI + BS) √∑ 2 ‚ñ∑ Calcul de l'√©l√©ment du milieu
6:        si milieu √ó milieu > n alors
7:            retourner racine(n, BI, milieu) ‚ñ∑ Recherche dans la partie gauche
8:        sinon
9:            retourner racine(n, milieu, BS) ‚ñ∑ Recherche dans la partie droite
10:        fin si
11:     fin si
12: fin fonction
```




La fonction suivante retourne la racine carr√©e enti√®re de n √† condition que celle-ci soit comprise entre $B_I$ et $B_S ‚Äì 1$. L‚Äôappel `racine(n,1,n)` retournera donc la racine carr√©e enti√®re de $n$. On suppose que $n$ est sup√©rieur ou √©gal √† 2 et que $B_I$ est strictement inf√©rieur √† $B_S$.


````python
def racine(n, BI, BS):
    """
    Calcule la racine carr√©e enti√®re d'un nombre.

    Param√®tres:
    - n : entier naturel, le nombre dont on veut calculer la racine carr√©e enti√®re
    - BI : entier naturel, borne inf√©rieure de l'intervalle de recherche
    - BS : entier naturel, borne sup√©rieure de l'intervalle de recherche

    Retour:
    - entier naturel : racine carr√©e enti√®re de n
    """
    if BS == BI + 1:
        return BI  # Si l'intervalle est r√©duit √† sa plus simple expression, BI est la racine carr√©e enti√®re
    else:
        milieu = (BI + BS) // 2  # Calcul de l'√©l√©ment du milieu

        if milieu * milieu > n:
            return racine(n, BI, milieu)  # Recherche dans la partie gauche
        else:
            return racine(n, milieu, BS)  # Recherche dans la partie droite
````

````python
# Exemple d'utilisation :
nombre = 50  # On veut calculer la racine carr√©e enti√®re de 50
resultat = racine(nombre, 0, nombre + 1)
print("Racine carr√©e enti√®re de", nombre, ":", resultat)

# Un autre exemple:
nombre = 18  # On veut calculer la racine carr√©e enti√®re de 50
resultat = racine(nombre, 0, nombre + 1)
print("Racine carr√©e enti√®re de", nombre, ":", resultat)
````
